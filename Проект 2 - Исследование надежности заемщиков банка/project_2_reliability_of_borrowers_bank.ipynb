### Шаг 2.1 Заполнение пропусков

Убдеимся ещё раз, в каких столбцах есть пропуски, а так же узнаем их количество:

```python
df.isna().sum() # кол-во пропусков 
```

Мы видим какое количество пропусков присутствует в столбцах `days_employed` и `total_income`.  
Доля пропусков составляет более 10%, что не позволяет нам просто не обращать внимание на данные пропуски.
Причиной появления пропусков в данных может являться как технический фактор, так и человеческий. 

На данном этапе мы не будем трогать столбец `days_employed`, так как при первичном анализе обнаружили в нем отрицательные значения. А вот пропущенные значения столбца `total_income` мы можем заменить медианным значением. 

Медианное значение, в данном случае предпочтительней среднего арифмитического, потому что допустимый разброс между зарплатами заёмщиков может быть достаточно существенным.

### Шаг 2.2 Проверка данных на аномалии и исправления.

Проверим данные на аномалии. Попытаемся с помощью метода `head()` и `tail()` ухватить диапазон побольше, чтобы зацепиться глазом за какие-либо нереальные значения. 

```python
df.head(50) # первые 50 строчек таблицы df
```

```python
df.tail(50) # последние 50 строчек таблицы df
```

Проверим уникальное количество значений в столбце `children`:

```python
df['children'].value_counts() # проверка уникальных значений и их количества
```

Изучив выведенные данные, мы замечаем, что в столбце `days_employed` присутствуют преимущественно отрицательные значения, есть пропуски, а в строках, которые относятся к пенсионерам - нереально большие значения в днях для трудового стажа. В столбце `children` есть одно отрицательное значение (чего быть не может) и вариант с достаточно большим количеством детей.

**Выводы**

* нужно заменить все отрицательные значения на положительные;
* скорректировать трудовой стаж пенсионеров
* заполнить пропуски медианными значениями (так же по причине того, что имеет место быть большой разброс).


Исправим отрицательные значения на положительные в столбце `days_employed` и `children`:

```python
df['days_employed'] = df['days_employed'].abs() # меняем отрицательные значения на положительные
```

```python
df.head(10) # проверяем замену
```

```python
df['children'] = df['children'].abs() # меняем отрицательные значения на положительные
```

Значения в столбце `children`, где у семей по 20 детей, выглядят маловероятными, поэтому предположим, что в данных ошибка в виде лишнего нуля, учитывая что из-за небольшого значения, это мало на что повлияет. 

```python
# с помощью логической индексации все значения в колонке children заменим на 2
df.loc[(df.children == 20), 'children'] = 2
```

```python
df['children'].value_counts() # проверяем уникальные значения
```

Теперь мы имеем положительные значения на месте отрицательных и скоректированные значения по количеству детей в семье.

Так как огромное значение в стаже работы пенсионеров не удается переконвертировать во что-то разумное, предлагаю следующий метод. Взять за основу то, что каждый пенсионер работает с 20-ти лет (с погрешностью такой, что могли быть перерывы без работы), отнять данную цифру от нынешнего возраста и умножить на 365 дней в году.

```python
df.loc[df['income_type'] == 'пенсионер', 'days_employed'] = df.loc[df['income_type'] == 'пенсионер', 'dob_years'].sub(20).mul(365)
# путем логического подхода, заполняем пустые значения и некорректные данные в трудовом стаже пенсионеров.
```

```python
df.head(20)
```

Заменим оставшиеся пропуски на медианное значение в столбце `days_employed`:

```python
df['days_employed'] = df['days_employed'].fillna(df['days_employed'].median()) # замена пропусков на медианное значение
```

```python
df.isna().sum() 
```

Теперь мы видим, что больше у нас нет пустых значений.

### Шаг 2.6. Категоризация дохода.

Для дальнейшего анализа сделаем категаризацию данных по уровню дохода, где:

- "Е" - от 0 до 30 000,
- "D" - от 30 001 до 50 000,
- "C" - от 50 001 до 200 000,
- "B" - от 200 001 до 1 000 000,
- "A" - от 1 000 001 и выше.

Вынесем данные значения в отдельный столбец под названием `total_income_category`.

```python
def income_level(income): # с помощью функции выставим условия для разбивки на категории по уровню дохода
    if income <= 30000:
        return 'E'
    elif (income >= 30001) and (income <= 50000):
        return 'D'
    elif (income >= 50001) and (income <= 200000):
        return 'C'
    elif (income >= 200001) and (income <= 1000000):
        return 'B'
    else:
        return 'A'
```

```python
df['total_income_category'] = df['total_income'].apply(income_level) # применяем функцию методом apply к столбцу 'total_income' 
                                                      # и заносим полученные данные в новый столбец 'total_income_category'
```

```python
df.sample(10) # выведем 10 случайных строк
```

```python
round(df['total_income_category'].value_counts(normalize=True).mul(100).to_frame('% от общего кол-ва'),2) 
# категории заемщиков по уровню дохода в процентах
```

Видим, что большая часть заемщиков (около 75 %) относится к так называемому среднему классу "С".

### Шаг 2.7. Категоризация целей кредита.

Создадим новый столбец `purpose_category`, в котором систематизируем цели кредита в более компактные и понятные категории:

- 'проведение свадьбы'
- 'операции с автомобилем' 
- 'получение образования'
- 'операции с недвижимостью'

В первую очередь посмотрим и изучим значения столбца `purpose`, чтобы понять, какие значения можно определить в одну категорию.

```python
df['purpose'].unique() # изучаем уникальные значения столбца
```

Изучив уникальные значения столбца `purpose`, отфильтровали повторяющиеся фрагменты строк, чтобы в дальнейшем с помощью функции и метода `apply` создать столбец `purpose_category` с новыми категориями.

```python
def purpose_credit (pur_list): # функция, помогающая категоризировать цели кредита
    if 'свадьб' in pur_list:
        return 'проведение свадьбы'
    elif 'автомобил' in pur_list:
        return 'операции с автомобилем'
    elif 'образован' in pur_list:
        return 'получение образования'
    else:
        return 'операции с недвижимостью'
    
```

```python
df['purpose_category'] = df['purpose'].apply(purpose_credit) # методом apply применяем функцию и создаем новый столбец
```

```python
df.sample(10) # проверяем 10 строк рандомно, новый столбец создан, категории соответствуют.
```

```python
round(df['purpose_category'].value_counts(normalize=True).mul(100).to_frame('% от общего кол-ва'),2)
# категории по целям кредита в процентах
```

Большая часть кредитных целей (50,37%) заемщиков приходятся на операции с недвижимостью, меньшая часть (10,88 %) - на проведение свадеб, что говорит о том, что надежда есть, несмотря на то, что получение образования (18,69 %) занимает предпоследнее место.

В связи с тем, что мы распределили стобцы `purpose` и `total_income` на категории в новых столбцах, их можно удалить, чтобы таблица выглядела более компактно:

```python
df = df.drop(columns=['purpose', 'total_income']) # удаление лишних столбцов
df.sample(5) 
```
